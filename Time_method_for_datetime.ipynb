```python
import numpy as np
import pandas as pd
from datetime import datetime as dt
```

```python
help(dt)
```

```python
myyear = 2024
mymonth= 12
myday=7
myhour = 20
myminute=40
mysecond = 10
```

```python
mydatetime = dt(myyear,mymonth,myday,myhour,myminute,mysecond)
mydatetime
```

```python
mydatetime.year
```

```python
myser = pd.Series(['Dec 12, 2024', '2001-12-22',None])
myser
```

```python
mydateseries = pd.to_datetime(myser,format='mixed')#it will give in yyyy-mm-dd format
mydateseries
```

```python
obvi_eruope_date = '31-12-2024'# dd-m-yyyy
pd.to_datetime(obvi_eruope_date)
```

```python
eurodate2 = '10-12-2024' # in europe its 10th dec 2024 but in america they might think its 12th oct 2024, and pandas uses american format
pd.to_datetime(eurodate2)
```

```python
# to deal with this case we mention if first one is day or not
pd.to_datetime(eurodate2, dayfirst=True)
```

```python
style_date = '22---Dec---2001'
pd.to_datetime(style_date, format= '%d---%b---%Y') # %b because i have written mnth name in word
```

```python
custom_date = '22nd of december 2024' # pandas is smart enough to understand this date formt
pd.to_datetime(custom_date)
```
