### **Useful Methods in Pandas**

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('tips.csv')
```

```python
df
```

```python
# Lets try to grab the last the last four digit of CC Number
def last_four(num):
    return str(num)[-4:]
```

```python
# lets try to pass some number and check
last_four(df['CC Number'][0])
```

```python
df['CC Number'].apply(last_four)
```

```python
# lets try to create a payment/transaction id column
df['transaction'] = df['day'] + df['CC Number'].apply(last_four)
df['transaction']
```

```python
df
```

```python
df['tip'].max()
```

```python
def dollar_sign(num):
    if num < 10:
        return "$"

    elif num >= 10 and num < 30:
        return "$$"

    else:
        return "$$$"
```

```python
df['total_bill'].apply(dollar_sign)
# df['total_bill'].apply(dollar_sign).head(20)
```

```python
# using apply method with multiple columns
#tip percentage
def quality_of_tip(total_bill, tip):
    if tip/total_bill > 0.25:
        return "Generous"
    else:
        return "other"
```

```python
# df['tip quality'] =df[['total_bill', 'tip']].apply(quality_of_tip)
# df['tip quality'] = df[['total_bill','tip']].apply(lambda df: quality_of_tip(df['total_bill'],df['tip']),axis = 1 )
```

```python
df
```

```python
df['Quality'] = np.vectorize(quality_of_tip)(df['total_bill'],df['tip'])
```

```python
df
```

# **GROUP BY OPERATION**

```python
import numpy as np
import pandas as pd
```

```python
df3 = pd.read_csv("mpg.csv")
```

```python
df3
```

```python
df3.groupby('model_year').mean(numeric_only=True)
```

```python
df3.groupby(['model_year','cylinders']).mean(numeric_only=True)
```

```python
df3.groupby(['model_year','cylinders']).mean(numeric_only=True)['mpg']
```

```python
df3.groupby('model_year').describe()
```

```python
df3.groupby('model_year').describe()['mpg']
```

```python
df3.groupby('model_year').describe().transpose()
```

```python
year_cyl = df3.groupby(['model_year','cylinders']).mean(numeric_only = True)
year_cyl
```

```python
year_cyl.index.levels
```

```python
year_cyl.index
```

```python
year_cyl.loc[(70,4)]
```

```python
four_cyl_year = year_cyl.xs(key=4, level='cylinders')
four_cyl_year
```

```python
# cyl_year6and8 = year_cyl.xs(key=(6,8), level='cylinders') # this will give error because we can only put one value as key
# cyl_year6and8

# i want cylinder 6 and 8 for every model year
cyl_year6and8 = df3['cylinders'].isin([6,8])
df3[cyl_year6and8]
```

```python
df3[cyl_year6and8].groupby('model_year').mean(numeric_only = True)
```

```python
year_cyl.swaplevel()

# swapping levels , now cylinders will come first and model_year second
```
