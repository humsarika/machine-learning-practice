## **Matplotlib** - Implementing figures and axes

**Note**:
*   if we are running .py scripts then to display the plot we need to write plt.show()




```python
a = np.linspace(0,10,11)
a
```

```python
b = a**3
b
```

```python
x = np.arange(0,10)
x
```

```python
y= 2*x
y
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y);
# to get rid this message we can either add a semocolon at the end of 'axes.plot(x,y)' or we can write plt.show()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b);
```

```python
type(fig)
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.set_xlim(0,8)
axes1.set_ylim(0,1000)
axes1.set_xlabel('X axis')
axes1.set_ylabel('Y axis')
axes1.set_title('Power of 4')
axes1.plot(a,b);

axes2 = fig.add_axes([0.2,0.4,0.4,0.4])
axes2.set_xlim(7,8)
axes2.set_ylim(400,600)
axes2.set_xlabel('Zoomed in X axis')
axes2.set_ylabel('Zoomed in Y axis')
axes2.set_title('zoomed in plot')
axes2.plot(a,b);
```



### **Matplotlib figure objects**
figure parameters

```python
fig = plt.figure(figsize = (10,4), dpi = 200)
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b);
plt.savefig("power_of_4.png")
```

### **Note**: when we are downloading it or trying to open this plot after saving/downloading then the axis is not visible.
To resolve this we set attribute bbox_inches = 'tight'.

without bbox_inches='tight'
power_of_4.png

with bbox_inches='tight'

power_of_4 (1).png

```python
fig = plt.figure(figsize = (10,4), dpi = 200)
axes = fig.add_axes([0,0,1,1])
axes.plot(a,b);
plt.savefig("power_of_4.png", bbox_inches ='tight')
```

```python
x = np.arange(10)
x
```

```python
y = 2*x
y
```

```python
a= np.linspace(0,20,11)
a
```

```python
b = a**3
b
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(x,y)
axes[1].plot(a,b)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[1].plot(a,b)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_title('[0][0]')
axes[1][1].plot(a,b)
axes[1][1].set_title('[1][1]')

axes[0][1].set_title('[0][1]')
axes[1][0].set_title('[1][0]')
plt.tight_layout()
```

```python
help(plt.subplots_adjust)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)

plt.suptitle('Main Title')# super title
axes[0][0].plot(x,y)
axes[0][0].set_title('[0][0]')
axes[1][1].plot(a,b)
axes[1][1].set_title('[1][1]')

axes[0][1].plot(x,y)
axes[0][1].set_title('[0][1]')
axes[1][0].plot(a,b)
axes[1][0].set_title('[1][0]')

plt.subplots_adjust(wspace=0.4, hspace=0.4)

fig.savefig('subplot-example.png')
```

## **Matplotlib Styling- Legends**
---

[Matplotlib line markers](https://matplotlib.org/stable/api/markers_api.html)


- [Practice marker in this cell](https://colab.research.google.com/drive/1xoyy4_2xNTTDYpekTZjM6zXqdQqn0liF#scrollTo=dwsCXXOdeE5t&line=5&uniqifier=1)





```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x, label ='x vs x');
axes.plot(x,x**2, label ='x vs x**2');
axes.plot(x,2**x,label ='x vs 2**x');
axes.legend()
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x, label ='x vs x');
axes.plot(x,x**2, label ='x vs x**2');
axes.plot(x,2**x,label ='x vs 2**x');
axes.legend(loc = 9)#by default its 0
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x, label ='x vs x');
axes.plot(x,x**2, label ='x vs x**2');
axes.plot(x,2**x,label ='x vs 2**x');
axes.legend(loc =(0.4,-0.2))
# axes.legend(loc =(1.1,0.7))
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x, label ='x vs x', color='#db255c', linestyle=':');
axes.plot(x,x**2, label ='x vs x**2', color='#4bb4c4',ls='--');
axes.plot(x,x*5,label ='x vs 2**x', color='green', ls = '-.', linewidth = 3);
axes.legend(loc =(0.4,-0.2))
#supported values for ls )linestyle are '-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x, label ='x vs x', color='#db255c', lw=1, linestyle=':', marker='.'); #we can use ms/markersize to edit the size of marker
axes.plot(x,x**2, label ='x vs x**2', color='#4bb4c4', linewidth=3, ls='--');
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x, label ='x vs x', color='#db255c', lw=1, linestyle=':', marker='.', ms=10, markeredgewidth=1, markerfacecolor='yellow', markeredgecolor ='black');
axes.plot(x,x**2, label ='x vs x**2', color='#4bb4c4', linewidth=3, ls='--');
```
