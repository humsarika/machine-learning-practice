

>Subplot functionality



The plt.subplots() call allows us to easily create figure and axes objects in side by side formations.

The plt.subplots() commands returns a tuple containing the figure canvas and then numpy array holding the axes objects.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10) 
y = x*2
```

```python
fig , axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(a,b)
axes[1].plot(x,y)
```

```python
type(axes)
```

Axes is no longer a matplotlib axes object it's actually a numpy array.

It means it has a shape.

```python
axes.shape
```

```python
axes
```

```python
fig , axes = plt.subplots(nrows = 3, ncols = 1) # this is again a 1 dimensional
for i in axes:
  i.plot(x,y)
#  so we can do the the ploting of diffrent data sets or of the same 
# (for practice only I'm using a for loop to plot the the same data in every plot)
# we can see that there is overlaping in data in plots to fix this we write a command in last
plt.tight_layout()
```

```python
fig , axes = plt.subplots(nrows = 2, ncols = 2,dpi=100) # now it has became 2 dimensional so  to grab any plot we use [][] 2d array calls
axes[0][0].plot(x,y)
axes[0][0].set_ylabel('y')
axes[0][0].set_xlabel('x')
axes[0][0].set_title('Title 0,0')

axes[0][1].plot(a,b)
axes[0][1].set_ylabel('y')
axes[0][1].set_xlabel('x')
axes[0][1].set_title('Title 0,1')

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('y')
axes[1][0].set_xlabel('x')
axes[1][0].set_title('Title 1,0')

axes[1][1].plot(a,b)
axes[1][1].set_ylabel('y')
axes[1][1].set_xlabel('x')
axes[1][1].set_title('Title 1,1')
fig.suptitle('4 plots in 1 figure ',fontsize = 16)
# plt.tight_layout()
fig.subplots_adjust(top=0.8 ,wspace = 1,hspace = 1)
# def subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=None)
fig.savefig('4plots.jpg',bbox_inches = 'tight')
```

```python
plt.plot(m,E,color='red')
plt.title('E = mc^2')
plt.xlabel('m in g')
plt.ylabel('E in j')
plt.xlim(0,10)
# if we want a log scale 
plt.yscale("log")
# plt.xscale("log")
plt.grid(which = 'both',axis = 'y')
# supported values for which are 'major', 'minor', 'both'
# supported values for axis are 'x' , 'y, 'both'
```

```python
labels = ['1 mo','3 mo','6 mo','1 yr','2 yr','3 yr','5 yr','7 yr','10 yr','20 yr','30 yr']
july16_2007 = [4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(labels , july16_2007,label = 'july 16 2007')
ax.plot(labels , july16_2020,label = 'july 16 2020')
plt.legend(loc=(1.03,0.4))
```

```python
fig , axes = plt.subplots(nrows = 2, ncols = 1,figsize= (10,6))
axes[0].plot(labels , july16_2007)
axes[0].set_title('july 16 2007')
axes[1].plot(labels , july16_2020)
axes[1].set_title('july 16 2020')

plt.tight_layout()
```

```python
fig, ax1 = plt.subplots(figsize = (10,8))

ax1.spines['left'].set_color('blue')
ax1.spines['left'].set_linewidth(3)

ax1.plot(labels , july16_2007, lw=2, color="blue")
ax1.set_ylabel(r"july 16 2007", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax1.spines['right'].set_color('red')
ax1.spines['right'].set_linewidth(3)

ax2 = ax1.twinx()
ax2.plot(labels , july16_2020, lw=2, color="red")
ax2.set_ylabel(r"july 16 2020", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
    # The Axes.get_yticklabels() function in
    # axes module of matplotlib library is used 
    # to return the y ticks as a list of locations.
```
