**Group By Operations**                                                         
A groupby() operation allows us to examine data on a per category basis.


```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('mpg.csv')
```

```python
df
```

```python
df['model_year'].unique()
```

```python
df['model_year'].value_counts()
```

```python
df.groupby('model_year').mean()
# we have a name column but it returns only numeric column
```

```python
df.groupby('model_year').mean()['mpg']
```

```python
df.groupby(['model_year','cylinders']).mean()
```

```python
# df.groupby(['model_year','cylinders','origin']).mean()
```

```python
df.groupby(['model_year','cylinders']).mean().index
```

```python
df.groupby(['model_year','cylinders']).mean()['mpg']
```

```python
df.groupby('model_year').describe()
```

```python
# df.groupby('model_year').describe().transpose()
# more readable
```

```python
year_cyl = df.groupby(['model_year','cylinders']).mean()
```

```python
# year_cyl
year_cyl.index.names
```

```python
year_cyl.index.levels
```

```python
# working with index
year_cyl.loc[[70,80]]
```

```python
year_cyl.loc[(70,4)]
```

```python
year_cyl.xs(key = 70,level = 'model_year') # cross section
```

```python
year_cyl.xs(key =4 , level = 'cylinders')
# year_cyl.xs(key =[6,8] , level = 'cylinders') <---this won't work
```

```python
df[df['cylinders'].isin([6,8])]
```

```python
df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean()
# we won't do this by cross section (.xs) so we do this by isin


# filter out  data first then group by any column 
```

```python
year_cyl.sort_index(level = 'model_year', ascending=False)
# sorting by model year in descending
```

```python
year_cyl.swaplevel()
```
