**Combining DataFrames - merge**

The `.merge()` method takes in a key argument labeled below

There are  3 main ways to of merging tables together using the `how` parameter :

```
*  Inner - the result will be the set of records that match in the both tables. (intersection of both the data)
*  Outer - 
*  Left or Right
```


>Inner merge










````python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bob','Claire','David']},index=[1,2,3,4])
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','John','Bob']},index=[1,2,3,4])
````

````python
registrations
````

````python
logins
````

````python
# help(pd.merge)
````

````python
pd.merge(registrations,logins,how = 'inner',on ='name') # switching the orders doesn't matter
````



> Left or Right merge



````python
pd.merge(registrations,logins,how = 'left',on ='name') 
# every single name in the left table will be present in the results along with their corresponding column values between the two tables.
# claire an david were not actually present in the login table means they dont have a log id associated with them , so that null values were filled in the log_id column.
````

````python
pd.merge(registrations,logins,how = 'right',on ='name')
#  every single name in the right table will be present in the results along with their corresponding column values between the two tables.
# Xavier and john were not actually present in the registrations table means they dont have a reg_id associated with them , so that null values were filled in the reg_id column.
````



> Outer merge



````python
pd.merge(registrations,logins,how = 'outer',on ='name')
#  all the names of the both the table will be present in the results along with their corresponding column values between the two tables. 
#  but missing values from log_id column and reg_id column will filled with nan values
````
