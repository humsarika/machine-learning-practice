```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Sales_Funnel_CRM.csv')
```

```python
df.head()
```

```python
Licenses = df[['Company','Product','Licenses']]
```

```python
Licenses
```

```python
pd.pivot(data = Licenses, index = 'Company', columns = 'Product', values = 'Licenses')
```

```python
df.groupby('Product').sum('Sale Price')
```

```python
df.groupby(['Company', 'Product']).sum('Sale Price')
```

```python
df.groupby(['Company']).sum('Sale Price')
```

```python
pd.pivot_table(df,index='Company', aggfunc='sum')
```

```python
# lets do this by pivot tables
# finding total sale price of these products
# pd.pivot_table(df,index='Company', columns='Product', values=['Sale Price'])
pd.pivot_table(df,index='Company', aggfunc='sum', values=['Sale Price','Licenses'])
```

```python
pd.pivot_table(df,index=['Account Manager', 'Contact'], aggfunc='sum', values=['Sale Price','Licenses'])
```

```python
pd.pivot_table(df,index=['Account Manager', 'Contact'], columns ='Product', aggfunc='sum', values=['Sale Price','Licenses'], fill_value=0)
```
